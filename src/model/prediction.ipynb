# Importing Important Libraries

```python
import pickle
import pandas as pd
```

# Loading model

```python
with open("classification_model.pickle","rb" ) as file:
    model = pickle.load(file)
```

# Loading dataset

```python
df = pd.read_csv("train_dataset.csv")
df.head(3)
```

# Seperating data set for prediction

```python
bucket_predict = df.sample(100000, random_state=42)
```

# Seperating true label

```python
true_classify = bucket_predict["AQI_Bucket"]
print(true_classify)
```

```python
bucket_predict.drop(columns="AQI_Bucket", inplace=True)
```

```python
bucket_predict
```

# Prediction

```python
pred_classify = model.predict(bucket_predict)
```

# Evaluation

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report
accu = accuracy_score(y_true=true_classify, y_pred=pred_classify)
print(f"The accuracy of model is {round(accu,3) * 100 }")
```

```python
f1 = f1_score(y_true=true_classify, y_pred=pred_classify, average=None)
print(f"The accuracy of model is {f1}")
```

```python
c_report = classification_report(y_true=true_classify, y_pred=pred_classify)
print(f"The accuracy of model is {c_report}")
```
